# Exploratory Data Analysis (EDA)
It is the process to understand the data in depth and learn the data characteristics often with visual means.
We perfom EDA to explore, clean and filter the data before we apply any machine learning algorithms.
EDA is important cause:
1.Helps identify errors in data sets.
2.Gives a better understanding of the data set. 
3.Helps detect outliers or anomalous events.
4.Helps understand data set variables and the relationship among them.
